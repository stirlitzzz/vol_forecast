```python
def get_option_data(ticker, date):
    getMarketData = ivol.setMethod('/equities/eod/stock-opts-by-param')
    # Fetch call options data
    calls = getMarketData(
        symbol=ticker,
        tradeDate=date,
        dteFrom=0,
        dteTo=120,
        deltaFrom=0.2,
        deltaTo=0.8,
        cp='C'
    )

    # Fetch put options data
    puts = getMarketData(
        symbol=ticker,
        tradeDate=date,
        dteFrom=0,
        dteTo=120,
        deltaFrom=-0.8,
        deltaTo=-0.2,
        cp='P'
    )

    return calls, puts

def find_closest_by_expiry(df, delta_col='delta_call', targets=[0.25, 0.75], n=2):
    """
    For each expiration date, find the n closest strikes for each target delta.
    """
    result = []
    for expiry, group in df.groupby('expiration_date'):
        for target in targets:
            closest = (
                group
                .assign(delta_diff=(group[delta_col] - target).abs())
                .nsmallest(n, 'delta_diff')
            )
            result.append(closest)
    return pd.concat(result).drop_duplicates(subset=['price_strike', 'expiration_date'])


def find_nearest_strikes_around_spot(df, spot_col='underlying_price_call'):
    """
    For each expiration, find one strike below and one above the spot.
    """
    result = []
    for expiry, group in df.groupby('expiration_date'):
        spot = group[spot_col].iloc[0]  # assume same spot across expiry group
        below = group[group['price_strike'] <= spot].sort_values('price_strike').tail(1)
        above = group[group['price_strike'] >= spot].sort_values('price_strike').head(1)
        result.extend([below, above])
    return pd.concat(result)




def load_and_filter_day_ticker(ticker, date):
    # --- 1. Fetch data ---
    print(f"Fetching data for {ticker} on {date}")
    call_data,put_data = get_option_data(ticker, date)
    print(f"Fetched {len(call_data)} call options and {len(put_data)} put options")
    # --- 2. Filter out empty data ---
    
    #print(f"Call data: {call_data.head()}")
    #print(f"Put data: {put_data.head()}")
    if call_data.empty or put_data.empty:
        return None  # skip if anything is missing
    
    # --- 2. Merge on strike (assumes same expiration, maturity) ---
    #merged = pd.merge(call_data, put_data, on='strike', suffixes=('_call', '_put'))

    merged = pd.merge(
        call_data,
        put_data,
        on=['price_strike', 'expiration_date'],
        suffixes=('_call', '_put')
    )

    #print(f"Merged data: {merged.head()}") 
    # --- 3. Compute deltas if not present (or use if already there) ---
    # e.g., merged['delta_call'], merged['delta_put'] -- assumed provided
    
    # --- 4. Compute delta-weighted vol ---
    merged['vol_weighted'] = (
        merged['delta_call'] * merged['iv_call'] +
        (1 - merged['delta_call']) * merged['iv_put']
    )
    
    #print(f"Computed vol_weighted: {merged['vol_weighted'].head()}")
    # --- 5. Select strikes around key deltas (and ATM) ---
    def closest(df, col, target, n=2):
        return df.assign(diff=(df[col] - target).abs()).nsmallest(n, 'diff')
    
    #keep = find_closest_by_expiry(
    #    merged,
    #    delta_col='delta_call',
    #    targets=[0.25, 0.5, 0.75],
    #    n=2
    #)

    # 1. Delta targets per expiry
    delta_filtered = find_closest_by_expiry(merged)
    #print(f"Delta filtered data: {delta_filtered.head()}")

    # 2. ATM pair per expiry
    atm_pair = find_nearest_strikes_around_spot(merged)
    #print(f"ATM pair data: {atm_pair.head()}")

    # 3. Combine and deduplicate
    keep = pd.concat([delta_filtered, atm_pair]).drop_duplicates(subset=['price_strike', 'expiration_date'])
    return keep



#calls, puts = get_option_data('AAPL', '2021-12-16')

keep = load_and_filter_day_ticker('AAPL', '2021-12-30')
```

def extract_vol_features_minimal(df):
    spot = df['underlying_price_call'].iloc[0]
    df = df.copy()
    df['moneyness'] = df['price_strike'] / spot
    df['abs_diff'] = (df['price_strike'] - spot).abs()

    # --- 1. ATM vol via interpolation ---
    atm_pair = df.nsmallest(2, 'abs_diff').sort_values('price_strike')
    k1, k2 = atm_pair['price_strike'].values
    v1, v2 = atm_pair['vol_weighted'].values

    if k2 != k1:  # avoid divide by zero
        weight = (spot - k1) / (k2 - k1)
        atm_vol = v1 + weight * (v2 - v1)
    else:
        atm_vol = v1  # degenerate case: both strikes equal

    # --- 2. Upside slope (ATM → higher strike) ---
    up = df[df['price_strike'] > spot].sort_values('price_strike').head(1)
    slope_up = (
        (up['vol_weighted'].values[0] - atm_vol) /
        (up['moneyness'].values[0] - 1)
        if not up.empty else None
    )

    print(f'spot=')
    # --- 3. Downside slope (lower strike → ATM) ---
    down = df[df['price_strike'] < spot].sort_values('price_strike', ascending=False).head(1)
    slope_down = (
        (atm_vol - down['vol_weighted'].values[0]) /
        (1 - down['moneyness'].values[0])
        if not down.empty else None
    )

    return {
        'atm_vol': atm_vol,
        'slope_up': slope_up,
        'slope_down': slope_down
    }
